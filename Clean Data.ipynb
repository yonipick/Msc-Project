```
## Function to reduce the DF size
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2    
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

```
# missing values check method
def check_missing_values_in_df(df):
  """ This method gets a DataFrame and checks for missing values in it"""
  total = df.isnull().sum().sort_values(ascending = False)
  percent = (df.isnull().sum() / df.isnull().count() * 100).sort_values(ascending = False)
  missing_df_data = pd.concat([total, percent], axis=1, keys = ['Total', 'Percent'])  
  return missing_df_data
```

```
def drop_na(df):
  before_observations = df.shape[0]
  df.dropna(inplace=True)
  after_observations = df.shape[0]
  now = before_observations - after_observations
  print(f"Now the is {after_observations} observations, remove {now} that are {round(now/after_observations *100, 4) }% from the data")
  return df
```

```
def clean_text(text):
    text = re.sub(r"AI", "artificial intelligence", text)
    text = re.sub(r"A.I.", "artificial intelligence ", text)
    text = text.lower()
    text = re.sub(r"what's ", "what is ", text)
    text = re.sub(r"\'s ", " ", text)
    text = re.sub(r"\'ve ", " have ", text)
    text = re.sub(r"can't ", "can not ", text)
    text = re.sub(r"n't ", " not ", text)
    text = re.sub(r"i'm ", "i am ", text)
    text = re.sub(r"im ", "i am ", text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub(r"[^a-z]", ' ', text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
        
    return text
```

```
def remove_blanks(df,col=None):
  X = df.loc[:,[col]]
  blanks = []
  #(index, label, review)
  for i, item,  in X.itertuples():
    if item.isspace():
      blanks.append(i)
  df.drop(blanks,inplace=True)
  print(f"Number of blanks items {len(blanks)}")
  return df
```

```
def unite_colums(df):
  df['machine learning'] = df['ai'] | df['machine learning'] | df['deep learning']
  df['design'] = df['design'] | df['architecture'] | df['product design'] | df['ux']
  df['blockchain'] = df['blockchain'] | df['bitcoin'] | df['cryptocurrency'] | df['ethereum']
  df['writing'] = df['writing'] | df['blogging'] | df['creativity'] | df['medium']
  df['marketing'] = df['marketing'] | df['branding']
  df['art'] = df['art'] | df['college']
  df['programming'] = df['computer science'] | df['programming'] | df['python']
  df['data science'] = df['data science'] | df['data visualization']
  df['politics'] = df['politics'] | df['donald trump'] | df['economics'] | df['government'] | df['racism'] | df['trump']
  df['sports'] = df['sports'] | df['football'] | df['nba']
  df['life lessons'] = df['life lessons'] | df['happiness'] | df['inspiration']
  df['self improvement'] = df['self improvement'] | df['motivation'] | df['personal development'] 
  df['food'] = df['food'] | df['nutrition']
  df['love'] = df['love'] | df['poem'] | df['relationships']
  df['javascript'] = df['javascript'] | df['react'] 
  df['entrepreneurship'] = df['entrepreneurship'] | df['venture capital']
  df['web development'] = df['web development'] | df['web design'] 


  df = df.drop(labels=['ai', 'architecture', 'bitcoin', 'blogging', 'branding',
                       'college', 'computer science', 'creativity', 'cryptocurrency', 
                       'data visualization', 'deep learning', 'donald trump', 'economics',
                       'ethereum', 'football', 'government', 'happiness', 'inspiration',
                       'medium', 'motivation', 'nba', 'nutrition', 'personal development',
                       'poem', 'product design', 'python', 'racism', 'react', 'relationships',
                       'trump', 'ux', 'venture capital', 'web design'], axis=1)
  
  return df
```
